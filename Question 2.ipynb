```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_excel('Admission 2019 Data for ISB Campus.xlsx',header = 1)
pd.set_option('display.max_columns',None)
data
```

```python
df = pd.DataFrame(data)
```

```python
df.rename(columns = {'Ist': '1st_campus_pref', '2nd': '2nd_campus_pref','3rd': '3rd_campus_pref','4th': '4th_campus_pref'}, inplace = True)
```

```python
df.rename(columns = {'Ist.1': '1st_discpline_pref', '2nd.1': '2nd_discpline_pref','3rd.1': '3rd_discpline_pref','4th.1': '4th_discpline_pref'}, inplace = True)
```

```python
df.rename(columns = {'Board': 'Board_Matriculation', 'Degree': 'Degree_Matriculation','Total': 'Total_Matriculation','Obtained': 'Obtained_Matriculation'}, inplace = True)
```

```python
df.rename(columns = {'Board.1': 'Board_Intermediate', 'Degree.1': 'Degree_Intermediate','Total.1': 'Total_Intermediate','Obtained.1': 'Obtained_Intermediate'}, inplace = True)
```

```python
df.rename(columns = {'Total.2': 'Total_SAT_I','Obtained.2': 'Obtained_SAT_I'}, inplace = True)
```

```python
df.rename(columns = {'Total.3': 'Total_SAT_II','Obtained.3': 'Obtained_SAT_II'}, inplace = True)
```

```python
df.rename(columns = {'BBA': 'BBA_NU_Test_Marks','BS': 'BS_NU_Test_Marks'}, inplace = True)
```

```python
df.rename(columns = {'BBA/BS(AF)': 'BBA/BS(AF)_NU_Merit_Marks','BS.1': 'BS_NU_Merit_Marks'}, inplace = True)
```

```python
df.rename(columns = {'Campus.1': 'Campus_NU_Admission_Offer','Discipline.1': 'Discipline_NU_Admission_Offer'}, inplace = True)
```

```python
df.rename(columns = {'Campus.2': 'Campus_Admission','Discipline.2': 'Discipline_Admission'}, inplace = True)
```

```python
x = df.loc[df["Total_Intermediate"] > df["Obtained_Intermediate"]]
x
```

```python
y = pd.DataFrame(x)
```

```python
y["Degree_Intermediate"].isnull().sum()
```

```python
y.loc[x["DEGREE_STATUS"]=='Current']
```

```python
y['DEGREE_STATUS'].value_counts().plot(kind = 'bar', figsize = (10,10), color = 'y')

plt.ylabel("Total Number of Student")
plt.xlabel("DEGREE STATUS")
```

```python
y['Percentage'] = (y['Obtained_Intermediate']/y['Total_Intermediate'])*100
y['Percentage'].head()
```

```python
z = pd.DataFrame(y[y['Percentage'] > 50].groupby('DEGREE_STATUS')['Percentage'].count())
print(z)
z.plot.barh(figsize=(20,20),color='y')
```

```python
z = pd.DataFrame(y[y['Percentage'] < 50].groupby('DEGREE_STATUS')['Percentage'].count())
print(z)
z.plot.barh(figsize=(20,20),color='y')
```
